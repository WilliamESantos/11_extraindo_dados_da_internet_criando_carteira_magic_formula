```python
# Importar as bibliotecas:
# Pandas é biblioteca open-source e de uso gratuito, que fornece ferramentas para análise e manipulação de dados em Python.
import pandas as pd
# O Selenium é o principal paocote para fazer web scraping
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
# O webdriver é um gerenciados de driver ex. (corregir dados de drive para correção e conecção)
from webdriver_manager.chrome import ChromeDriverManager
```

### <font color='darkblue'> Entendendo Como Funcionam as Requisições na internet

```python
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
url = 'https://www.fundamentus.com.br/resultado.php'
driver.get(url)
```

### <font color='darkblue'> Lendo os Dados do Site Fundamentus

```python
local_tabela = '/html/body/div[1]/div[2]/table'
tabela = driver.find_element("xpath", local_tabela)
html_tabela = tabela.get_attribute('outerHTML')
tabela = pd.read_html(str(html_tabela), thousands = '.', decimal = ',')[0]
tabela
```

### <font color='darkblue'> Construindo a Tabela Final

```python
tabela = tabela.set_index("Papel")
tabela = tabela[['Cotação', 'EV/EBIT', 'ROIC', 'Liq.2meses']]
tabela
```

### <font color='darkblue'> Transformando o Texto em Números

```python
#tabela['EV/EBIT'] * tabela['ROIC']
tabela.info()
```

### <font color='darkblue'> Tratamento de Informação

```python
# Vamos transformar o ROIC "object" em número
# Vamos usar a função st,replace para fazer essa conversão de toda a tabela ou coluna
# Nesse caso vamos transformar apena a coluna ROIC em número, observar que o Brasil trabalha com (ponto e virgula)
```

```python
tabela['ROIC'] = tabela['ROIC'].str.replace('%', '')
tabela['ROIC'] = tabela['ROIC'].str.replace('.', '', regex=True)
tabela['ROIC'] = tabela['ROIC'].str.replace(',', '.', regex=True)  
tabela['ROIC'] = tabela['ROIC'].astype(float)
tabela
```

### <font color='darkblue'> Fazendo os Filtros e Criando os Rankings

Efetuamos o filtro com a coluna "Liquizes" maior que R$ 1.000.000,00

```python
tabela = tabela[tabela['Liq.2meses'] > 1000000]
tabela
```

```python
tabela = tabela[tabela['EV/EBIT'] > 0]
tabela = tabela[tabela['ROIC'] > 0]
tabela
```

### <font color='darkblue'> Observação Importante

Nós queremos o índice EV/EBIT o menor possível, e o ROIC maior possível
Ou seja, queremos a empresa mais barata (EV/BIT), com maior qualidade (ROIC) ao mesmo tempo

Seque a carteira encolhida das 10 melhores empresas para compra nessa data:

```python
tabela['ranking_ev_ebit'] = tabela['EV/EBIT'].rank(ascending = True)
tabela['ranking_roic'] = tabela['ROIC'].rank(ascending = False)
tabela['ranking_total'] = tabela['ranking_ev_ebit'] + tabela['ranking_roic']

tabela = tabela.sort_values('ranking_total')
tabela.head(10)
```
